```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from copy import deepcopy

data = pd.read_csv('crime-train.csv')
data_np = data.values
y = data_np[:,0]
X = data_np[:,1:]
N = len(data['ViolentCrimesPerPop'].values)
X2 = np.c_[np.ones(N), X]

w_init = np.zeros(X2.shape[1])




def Lasso_fit(X, y,lambda_,w):
    error_history = [0] * 1000000
    def soft_thresholding_operator(x, lambda_):
        if x  < -lambda_/2:
          return x + lambda_/2
        elif x > lambda_/2:
          return x - lambda_/2
        else:
          return 0
    
    i=0
    while(1):
      
        for j in range(0, len(w)):
            tmp_w = deepcopy(w)
            tmp_w[j] = 0.0
            r_j = y - np.dot(X, tmp_w)
            arg1 = np.dot(X[:, j], r_j)
            arg2 = lambda_
            #arg2 = (lambda_*X.shape[0])/2

            w[j] = soft_thresholding_operator(arg1, arg2)/np.sum((X[:, j]**2))
            print(j,arg1,arg2,w[j])
        if (np.max(np.abs(w-tmp_w))<10**-6):
            break
        
        error_history[i]=np.sum((r_j) ** 2)
        i=i+1
        
    return w, error_history,i


Final_w, error_history,i = Lasso_fit(X2,y,600,w_init)
```
